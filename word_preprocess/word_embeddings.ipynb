```python
import pandas as pd
```

```python
annot_df = pd.read_csv('/Users/jeremyhudsonchan/Dropbox/Files/Github_Repos/Text-Extraction-from-Images/preprocessed_data/annotations.csv')
```

```python
annot_df.head()
```

```python
# gensim import word2vec
from gensim.models import word2vec
from gensim.models import Word2Vec
word2vec_model = word2vec.Word2Vec(annot_df['utf8_string'], vector_size=10, window=5, min_count=5, workers=4)
# get all words in utf8_string column, convert them into vectors using word2vec, then update the column, if word not in vocabulary, ignore
def get_word2vec(word):
    try:
        return word2vec_model.wv[word]
    except:
        return None

annot_df['utf8_string'] = annot_df['utf8_string'].apply(lambda x: [get_word2vec(word) for word in x])
```

```python
# save the dataframe to a csv file, print out the process time
import time
start_time = time.time()
annot_df.to_csv('annot.csv')
print("--- %s seconds ---" % (time.time() - start_time))
```
